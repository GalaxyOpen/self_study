```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.binary_location = "/usr/bin/google-chrome"

service = Service("/usr/bin/chromedriver")
driver = webdriver.Chrome(service=service, options=options)

query = "금융"
url = f"https://search.naver.com/search.naver?query={query}&where=news"
driver.get(url)

titles = []
try:
    wait = WebDriverWait(driver, 10)
    elements = wait.until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "span.sds-comps-text-type-headline1"))
    )
    titles = [el.text.strip() for el in elements[:10] if el.text.strip()]
except Exception as e:
    print(f"❌ 뉴스 제목 추출 실패: {e}")
finally:
    driver.quit()

print("📌 추출된 뉴스 제목 개수:", len(titles))
print(titles)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch.nn.functional as F
import torch

# ✅ 분석 대상 뉴스 제목들 (예시로 titles 변수 사용)
# titles = [ ... ] ← 이전 크롤링 코드에서 가져온 것 그대로 사용

# -------------------------
# 1. FinBERT (영문)
print("\n🟦 FinBERT (영문 금융 특화)")
tokenizer_en = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model_en = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
pipe_en = pipeline("sentiment-analysis", model=model_en, tokenizer=tokenizer_en)

for title in titles:
    result = pipe_en(title)[0]
    print(f"제목: {title}\n→ 감성: {result['label']} (score: {result['score']:.2f})\n")

# -------------------------
# 2. KR-FinBERT (한글 금융 특화)
print("\n🟩 KR-FinBERT (한글 금융 특화)")
tokenizer_kr = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")
model_kr = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC")
pipe_kr = pipeline("sentiment-analysis", model=model_kr, tokenizer=tokenizer_kr)

for title in titles:
    result = pipe_kr(title)[0]
    print(f"제목: {title}\n→ 감성: {result['label']} (score: {result['score']:.2f})\n")

# -------------------------
# 3. KcELECTRA (한글 일반 감성 분석) → 미세조정 X이므로 레이블 해석 필요
print("\n🟨 KcELECTRA (한글 일반, 감성 미세조정 X)")
tokenizer_kc = AutoTokenizer.from_pretrained("beomi/KcELECTRA-base")
model_kc = AutoModelForSequenceClassification.from_pretrained("beomi/KcELECTRA-base")
model_kc.eval()

for title in titles:
    inputs = tokenizer_kc(title, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model_kc(**inputs)
        probs = F.softmax(outputs.logits, dim=1)
        pred = torch.argmax(probs, dim=1).item()
        label = f"LABEL_{pred}"  # 사전학습 모델은 실제 레이블 없음
        print(f"제목: {title}\n→ 감성 예측 (미세조정 안됨): {label} (score: {probs[0][pred]:.2f})\n")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# 🔹 영어 금융용 FinBERT
tokenizer_en = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model_en = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
pipe_en = pipeline("sentiment-analysis", model=model_en, tokenizer=tokenizer_en)

# 🔹 한국어 금융용 KR-FinBERT
tokenizer_kr = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")
model_kr = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC")
pipe_kr = pipeline("sentiment-analysis", model=model_kr, tokenizer=tokenizer_kr)

# 🔹 한국어 범용 KcELECTRA (감성 분석 미세조정 안 됨)
tokenizer_kc = AutoTokenizer.from_pretrained("beomi/KcELECTRA-base")
model_kc = AutoModelForSequenceClassification.from_pretrained("beomi/KcELECTRA-base")
model_kc.eval()
```

```python
import pandas as pd
import torch.nn.functional as F
import torch

results = []

for title in titles:
    # ① FinBERT (영문)
    result_en = pipe_en(title)[0]
    label_en = result_en['label']
    score_en = round(result_en['score'], 2)

    # ② KR-FinBERT (한글 금융 특화)
    result_kr = pipe_kr(title)[0]
    label_kr = result_kr['label']
    score_kr = round(result_kr['score'], 2)

    # ③ KcELECTRA (한국어 일반, 미세조정 X → 그냥 예시용)
    inputs = tokenizer_kc(title, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model_kc(**inputs)
        probs = F.softmax(outputs.logits, dim=1)
        pred = torch.argmax(probs, dim=1).item()
        label_kc = f"LABEL_{pred}"
        score_kc = round(probs[0][pred].item(), 2)

    results.append({
        "뉴스 제목": title,
        "FinBERT 감성": label_en,
        "FinBERT 확률": score_en,
        "KR-FinBERT 감성": label_kr,
        "KR-FinBERT 확률": score_kr,
        "KcELECTRA 감성": label_kc,
        "KcELECTRA 확률": score_kc
    })

df = pd.DataFrame(results)
```

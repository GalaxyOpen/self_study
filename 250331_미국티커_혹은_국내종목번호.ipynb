```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import mplfinance as mpf
import koreanize_matplotlib
from matplotlib import font_manager, rc
```

```python
# 한글 폰트 적용
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

# 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 주식 데이터 불러오기 (각각 다른 종목)
df_kospi = fdr.DataReader('KS11', '2024-01-01', '2024-12-31')
df_kosdaq = fdr.DataReader('KQ11', '2024-01-01', '2024-12-31')
df_samsung = fdr.DataReader('005930', '2024-01-01', '2024-12-31')
df_nvidia = fdr.DataReader('NVDA', '2024-01-01', '2024-12-31')
```

```python
# 코스피 차트
mpf.plot(df_kospi, type='candle', style='yahoo', title='KOSPI KS11', mav=(5, 10))

# 코스닥 차트
mpf.plot(df_kosdaq, type='candle', style='yahoo', title='KOSDAQ KQ11', mav=(5, 10))

# 삼성전자 차트
mpf.plot(df_samsung, type='candle', style='yahoo', title='Samsung Electronics 005930', mav=(5, 10))

# 엔비디아 차트
mpf.plot(df_nvidia, type='candle', style='yahoo', title='Nvidia NVDA', mav=(5, 10))
```

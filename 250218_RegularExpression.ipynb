```python
### 정규표현식 실습
import re
r = re.compile("a.c")
r.search("kkk") # 아무런 결과도 출력되지 않는다.
```

```python
r.search('abc')
# <re.Match object; span=(0, 3), match='abc'>
# .은 위의 표에서 어떤 문자로도 인식 될 수 있으므로 abc라는 문자열은 a.c라는 정규표현식 패턴으로 매치가 됨.
```

```python
### ? 기호
r = re.compile("ab?c")
r.search("abbc") # 아무런 결과도 출력되지 않음
```

```python
r.search("ac")
# <re.Match object; span=(0, 2), match='ac'>
# b가 없는 것으로 판단하여 ac를 매치함. ? 없는 것으로 판단.
```

```python
r.search('abc')
# <re.Match object; span=(0, 3), match='abc'>
# b가 있는 것으로 판단하여 abc를 매치함. ? 있는것으로 판단.
```

```python
### * 기호
r = re.compile("ab*c")
r.search('a') # 아무런 결과도 출력되지 않음
```
